# Create a program that takes a student's numerical grade as input and prints a
#corresponding letter grade. Use if-else statements to classify the grade into
#categories such as "A," "B," "C," "D," or "F."

````python
student_grades=int(input("Enter your Marks"))
if student_grades >=90:
 print("A")
elif student_grades >=80 and student_grades<=89  :
  print("B")
elif student_grades >=70 and student_grades<=79  :
  print("C")
elif student_grades >=60 and student_grades<=69  :
  print("D")
else :
  print("F")
````
